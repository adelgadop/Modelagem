## Leapfrog

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v2 as imageio
import os
from matplotlib import animation, rc
from functions import *

# Condições do exercício
# ----------------------
Nx, dx, Nt, dt, U  = 101, 5000, 101, 100, 10, 
hora = round((Nt-1)*dt/3600,1)
x, _ = np.linspace(0, (Nx-1)*5000, Nx, retstep = True)
t, _ = np.linspace(0, (Nt-1)*dt, Nt, retstep = True)
CFL     = U*dt/dx       # Courant number  CFL = 1, dt = 500
ylabel =  "PM$_{10}$ ($\mu$g.m$^{-3}$)"
C = np.zeros((Nx, Nt))

for n in range(1,Nt):
    res = np.sin(2*np.pi/1800*n*dt)
    if res > 0 :
        C[51,n] = res
    if res < 0 :
        C[51,n] = 0

fig, ax = plt.subplots(1,2,figsize=(10,3), gridspec_kw={'wspace':.3})
im = ax[0].imshow(C.T, cmap="Reds", vmax=-.01, vmin=.5)
ax[0].set_ylim(0,Nt-1)
ax[0].set_ylabel("n")
ax[1].set_ylabel(ylabel)
ax[1].set_xlabel("Tempo (h)")
ax[1].set_title(f"t[{n}] = {round((n*dt/3600),1)} horas, $\Delta t$={dt} s", loc='left')
fig.colorbar(im, ax=ax[0],orientation="vertical", fraction=0.05, pad=0.09)
ax[1].plot(t/3600, C[51,:])
fig.savefig("fig/fonte.png", 
           dpi = 400, bbox_inches='tight', facecolor='w')
```
